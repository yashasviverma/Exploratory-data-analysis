# 2) What is the average number of words in movie titles between the years 2000-2005?

```python
df.head(2)
```

```python
df.original_title
```

```python
start_date = "2000-01-01"
end_date = "2005-12-31"
```

```python
a = df['release_date'] >= start_date
```

```python
b = df['release_date'] <= end_date
```

```python
between_2000_5 = a & b
```

```python
filter_data = df.loc[between_2000_5]
```

```python
new_data = filter_data
```

```python
new_data
```

```python
new_data.original_title
```

```python
count = new_data['original_title'].str.split().str.len()
```

```python
count.sort_index(inplace=False)
```

```python
count.mean()
```

ANS : The average number of words in movie titles between the years 2000-2005 are *2.7425569176882663*

# 5) What is the average runtime of movies in the year 2006?

```python
start_date = "2006-01-01"
end_date = "2007-01-01"
```

```python
start_d = df['release_date'] >= start_date
```

```python
end_d = df['release_date'] <= end_date
```

```python
dates_bet = start_d & end_d
```

```python
filter_d = df.loc[dates_bet]
```

```python
filter_d
```

```python
new_d = filter_d
```

```python
new_d.runtime.describe()
```

```python
mean_d = new_d["runtime"].mean()
```

```python
mean_d 
```

ANS : The average runtime of movies in the year 2006 is *104.89711934156378*
